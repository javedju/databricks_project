###Setup Batch Events
1. Create control schema
2. Create batch_events table
3. Insert an event record


```python
%sql
CREATE SCHEMA IF NOT EXISTS formula1.control
    MANAGED LOCATION 'abfss://formula1@dbextdlake.dfs.core.windows.net/control';
```

```python
%sql
CREATE TABLE IF NOT EXISTS formula1.control.batch_events
(
    batch_id INT,
    batch_date DATE
)
```
