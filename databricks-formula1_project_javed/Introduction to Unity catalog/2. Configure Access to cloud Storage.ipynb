####Create External Location

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS dbc_ext_dl_demo
    URL 'abfss://demo@dbextdlake.dfs.core.windows.net/'
    WITH (STORAGE CREDENTIAL `dbc-sc`)
    COMMENT 'External Location for Demo Container'
```
