# Set-up the project environment for Formula1 Project
1. Create External Location dbc_ext_dl_formula1-incr
2. Create Catalog formula1-incr
3. Create Schemas landing, bronze, silver and gold
4. Create Volume Files in the landing schema

```sql
%fs ls 'abfss://formula1-incr@dbextdlake.dfs.core.windows.net/landing'
```

```sql
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS dbc_ext_dl_formula1_incr
    URL 'abfss://formula1-incr@dbextdlake.dfs.core.windows.net/'
    WITH (STORAGE CREDENTIAL `dbc-sc`)
    COMMENT 'External Location for Formula1 Container'
```

```sql
SHOW CATALOGS;
```

```sql
CREATE CATALOG IF NOT EXISTS formula1_incr
   MANAGED LOCATION 'abfss://formula1-incr@dbextdlake.dfs.core.windows.net'
   COMMENT 'This is the main catalog for the formula1 project' ;
```

```sql
CREATE SCHEMA IF NOT EXISTS formula1_incr.landing;
CREATE SCHEMA IF NOT EXISTS formula1_incr.bronze
    MANAGED LOCATION 'abfss://formula1-incr@dbextdlake.dfs.core.windows.net/bronze';
CREATE SCHEMA IF NOT EXISTS formula1_incr.silver
    MANAGED LOCATION 'abfss://formula1-incr@dbextdlake.dfs.core.windows.net/silver';
CREATE SCHEMA IF NOT EXISTS formula1_incr.gold
    MANAGED LOCATION 'abfss://formula1-incr@dbextdlake.dfs.core.windows.net/gold'
```

```sql
%sql
SHOW SCHEMAS;
```

```sql
%sql
SELECT current_catalog();
```

```sql
USE CATALOG formula1_incr;
SHOW SCHEMAS;
```

```sql
%sql
CREATE EXTERNAL VOLUME IF NOT EXISTS  formula1_incr.landing.files
  LOCATION 'abfss://formula1-incr@dbextdlake.dfs.core.windows.net/landing'
```
