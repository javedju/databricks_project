```python
items = dbutils.fs.ls('/databricks-datasets/')
```

```python
folder_count = 0
file_count = 0

for item in items:
    if item.name.endswith('/'):
        folder_count += 1
    else:
        file_count += 1

print(f"Folders: {folder_count}")
print(f"Files: {file_count}")
```
