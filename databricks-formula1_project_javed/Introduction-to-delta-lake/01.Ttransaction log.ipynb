###Understanding Delta Lake Transaction Logs
Understanding the cloud storage directory structure for a delta lake

####1. Create Catalog and Schema for the Demo
**Catalog**: demo \
**Schema**: delta_lake

```sql
CREATE CATALOG IF NOT EXISTS demo
MANAGED LOCATION 'abfss://demo@dbextdlake.dfs.core.windows.net/'
```

```sql
CREATE SCHEMA IF NOT EXISTS demo.delta_lake
MANAGED LOCATION 'abfss://demo@dbextdlake.dfs.core.windows.net/delta_lake'
```

####2. Create a Delta Lake Table

```sql
CREATE TABLE IF NOT EXISTS demo.delta_lake.companies
(
    company_name STRING,
    founded_date DATE,
    country STRING
)
USING DELTA;
```

```sql
DESCRIBE EXTENDED demo.delta_lake.companies
```

```sql
INSERT INTO demo.delta_lake.companies
VALUES ('Databricks', '2008-01-01', 'USA'),
       ('Microsoft', '1975-04-04', 'USA'),
       ('Google', '1998-09-04', 'USA')
```
